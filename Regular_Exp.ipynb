```python
output=re.findall("no",data)
print(output)
```

```python
data.find("is")
```

```python
#to get all possible names in the data
output=re.findall("([A-Z][a-z]*)[\s-]([A-Z][a-z]*)",data)
print(output)
```

```python
output=re.findall("[\b(ha(ve|d)|a(m|re)|\bbe(en)?\b|d(o(ne)?|id)|s(ay|aid|een)|th(ink|ought))\b]",data)
print(output)
```

```python
#replaced first 100 space with 9
output=re.sub("\s", "9", data, 100)
print(output)
```

```python
output=re.search("The*",data)
print(output)
```

```python
output=re.search("[abc]",data)
print(output)
```

```python
#splits the data with respect to all spaces
output=re.split("\s",data)
print(output)
```
